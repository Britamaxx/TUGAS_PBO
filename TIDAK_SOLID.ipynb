```python
#BAGIAN O
# =========================================================
# NON SOLID - Open/Closed Principle (OCP)
# Tema: Sistem Perhitungan Biaya Pasien Rumah Sakit
#
# MASALAH:
# Setiap ada tipe pasien baru,
# kita harus mengubah method hitung_biaya().
#
# Ini melanggar OCP karena:
# - class TIDAK closed for modification
# - kode lama harus terus diubah
# =========================================================


class KalkulatorBiayaRumahSakit:

    # Method ini akan terus berubah
    # setiap ada tipe pasien baru
    def hitung_biaya(self, biaya_awal, tipe_pasien):

        # Pasien umum bayar normal
        if tipe_pasien == "umum":
            return biaya_awal

        # Pasien BPJS mendapat diskon 30%
        elif tipe_pasien == "bpjs":
            return biaya_awal * 0.7

        # Pasien VIP mendapat tambahan layanan
        elif tipe_pasien == "vip":
            return biaya_awal * 1.5

        # Pasien lansia mendapat diskon 20%
        elif tipe_pasien == "lansia":
            return biaya_awal * 0.8

        # Jika nanti ada pasien baru:
        # contoh:
        # pasien_asuransi
        # pasien_darurat
        # pasien_internasional
        #
        # maka kita HARUS mengubah method ini lagi
        #
        # Ini yang membuat kode jelek

        else:
            return biaya_awal


# =========================================================
# PROGRAM UTAMA
# =========================================================

# Membuat object kalkulator
kalkulator = KalkulatorBiayaRumahSakit()

# Contoh biaya rumah sakit
biaya = 1000000

# Menghitung biaya pasien BPJS
hasil_bpjs = kalkulator.hitung_biaya(biaya, "bpjs")

# Menghitung biaya pasien VIP
hasil_vip = kalkulator.hitung_biaya(biaya, "vip")

# Menampilkan hasil
print("=== NON SOLID OCP ===")
print("Biaya pasien BPJS :", hasil_bpjs)
print("Biaya pasien VIP  :", hasil_vip)


# =========================================================
# KESIMPULAN
# =========================================================
#
# Kekurangan kode ini:
#
# 1. Jika ada tipe pasien baru,
#    kita harus mengubah method hitung_biaya()
#
# 2. Method menjadi panjang dan sulit dibaca
#
# 3. Risiko bug lebih besar
#    karena kode lama sering diubah
#
# 4. Sulit maintenance
#
# Ini disebut:
# "Modification Heavy"
#
# =========================================================
```

```python
#BAGIAN L
# [SALAH] Semua dokter dipaksa bisa operasi

class Dokter:
    def _init_(self, nama):
        self.nama = nama

    def konsultasi(self):
        print(f"Dokter {self.nama} melakukan konsultasi.")

    def operasi(self):
        # Parent menganggap semua dokter bisa operasi
        print(f"Dokter {self.nama} melakukan operasi.")


class DokterSpesialisJantung(Dokter):
    def operasi(self):
        print(f"Dokter {self.nama} melakukan operasijantung.")


class DokterSpesialisGigi(Dokter):
    def operasi(self):
        print(f"Dokter {self.nama} melakukan operasi gigi.")


class DokterSpesialisKulit(Dokter):
    def operasi(self):
        print(f"Dokter {self.nama} melakukan tindakan bedah kulit ringan.")


class DokterPsikiater(Dokter):
    #  Tidak bisa operasi, tapi dipaksa punya method ini
    def operasi(self):
        raise Exception(f"ERROR: Dokter {self.nama} adalah psikiater, tidak bisa operasi!")


# =========================
# FUNGSI YANG MENGHARAPKAN SEMUA DOKTER BISA OPERASI
# =========================
def jadwal_operasi(dokter: Dokter):
    dokter.konsultasi()
    dokter.operasi()   # akan crash jika Psikiater


dr_anak = DokterSpesialisJantung("Andi")
dr_gigi = DokterSpesialisGigi("Budi")
dr_kulit = DokterSpesialisKulit("Citra")
dr_psikiater = DokterPsikiater("Dewi")

#  Aman
jadwal_operasi(dr_anak)
jadwal_operasi(dr_gigi)
jadwal_operasi(dr_kulit)

print("-" * 30)

#  ERROR (melanggar LSP)
jadwal_operasi(dr_psikiater)
```

```python
#BAGIAN I
class TenagaMedis:

    def operasi(self):
        raise NotImplementedError

    def resep_obat(self):
        raise NotImplementedError

    def cek_pasien(self):
        raise NotImplementedError


class Dokter(TenagaMedis):

    def operasi(self):
        print("Dokter melakukan operasi")

    def resep_obat(self):
        print("Dokter memberi resep obat")

    def cek_pasien(self):
        print("Dokter memeriksa pasien")


class Perawat(TenagaMedis):

    def operasi(self):
        raise NotImplementedError("Perawat tidak melakukan operasi")

    def cek_pasien(self):
        print("Perawat memeriksa pasien")

#Codingan ini salah karena seharusnya perawat hanya bisa mengecek pasien tidak bisa melakukan operasi
```

```python
#BAGIAN D
# --- CONTOH BURUK (MENGABAIKAN DIP) ---

class ObatKimia:
    """Kelas tingkat rendah (Konkret)"""
    def gunakan(self):
        return "Memberikan resep obat antibiotik."

class Dokter:
    """
    Kelas tingkat tinggi.
    MASALAH: Dokter ini bergantung langsung pada 'ObatKimia'.
    Jika Dokter ingin memberikan 'Terapi', kita harus mengubah kode kelas ini.
    """
    def __init__(self):
        # Ketergantungan langsung (Tight Coupling)
        self.metode = ObatKimia()

    def rawat_pasien(self):
        print(f"Dokter sedang: {self.metode.gunakan()}")

# Eksekusi
dr_budi = Dokter()
dr_budi.rawat_pasien()
```
