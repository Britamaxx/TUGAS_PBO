```python
#BAGIAN O
# =========================================================
# SOLID - Open/Closed Principle (OCP)
# Tema: Sistem Perhitungan Biaya Pasien Rumah Sakit
#
# SOLUSI:
# Gunakan inheritance + polymorphism
#
# Tujuan:
# Menambah tipe pasien TANPA mengubah kode lama
#
# Jadi:
# - Open for Extension
# - Closed for Modification
# =========================================================


# =========================================================
# BASE CLASS / INTERFACE
# =========================================================

class StrategiBiaya:

    # Method dasar
    # Akan dioverride oleh class turunannya
    def hitung_biaya(self, biaya_awal):
        raise NotImplementedError


# =========================================================
# STRATEGI PASIEN UMUM
# =========================================================

class PasienUmum(StrategiBiaya):

    # Pasien umum bayar normal
    def hitung_biaya(self, biaya_awal):
        return biaya_awal


# =========================================================
# STRATEGI PASIEN BPJS
# =========================================================

class PasienBPJS(StrategiBiaya):

    # Pasien BPJS mendapat diskon 30%
    def hitung_biaya(self, biaya_awal):
        return biaya_awal * 0.7


# =========================================================
# STRATEGI PASIEN VIP
# =========================================================

class PasienVIP(StrategiBiaya):

    # Pasien VIP mendapat layanan tambahan
    def hitung_biaya(self, biaya_awal):
        return biaya_awal * 1.5


# =========================================================
# STRATEGI PASIEN LANSIA
# =========================================================

class PasienLansia(StrategiBiaya):

    # Pasien lansia mendapat diskon 20%
    def hitung_biaya(self, biaya_awal):
        return biaya_awal * 0.8


# =========================================================
# KELAS KALKULATOR
# =========================================================

class KalkulatorRumahSakit:

    # Constructor menerima object strategi
    def __init__(self, strategi):
        self.strategi = strategi

    # Method ini TIDAK perlu diubah lagi
    def hitung(self, biaya_awal):
        return self.strategi.hitung_biaya(biaya_awal)


# =========================================================
# PROGRAM UTAMA
# =========================================================

biaya = 1000000


# ---------------------------------
# Pasien BPJS
# ---------------------------------

bpjs = KalkulatorRumahSakit(PasienBPJS())

hasil_bpjs = bpjs.hitung(biaya)

print("=== SOLID OCP ===")
print("Biaya pasien BPJS :", hasil_bpjs)


# ---------------------------------
# Pasien VIP
# ---------------------------------

vip = KalkulatorRumahSakit(PasienVIP())

hasil_vip = vip.hitung(biaya)

print("Biaya pasien VIP  :", hasil_vip)


# ---------------------------------
# Pasien Lansia
# ---------------------------------

lansia = KalkulatorRumahSakit(PasienLansia())

hasil_lansia = lansia.hitung(biaya)

print("Biaya pasien Lansia :", hasil_lansia)


# =========================================================
# CONTOH TAMBAH FITUR BARU
# =========================================================
#
# Misalnya nanti ada:
# Pasien Asuransi
#
# Kita TIDAK perlu mengubah
# class KalkulatorRumahSakit
#
# Cukup buat class baru saja
#
# =========================================================


class PasienAsuransi(StrategiBiaya):

    # Diskon 40%
    def hitung_biaya(self, biaya_awal):
        return biaya_awal * 0.6


# Menggunakan fitur baru
asuransi = KalkulatorRumahSakit(PasienAsuransi())

hasil_asuransi = asuransi.hitung(biaya)

print("Biaya pasien Asuransi :", hasil_asuransi)


# =========================================================
# KESIMPULAN
# =========================================================
#
# Kelebihan kode SOLID:
#
# 1. Tidak perlu mengubah kode lama
#
# 2. Mudah tambah fitur baru
#
# 3. Risiko bug lebih kecil
#
# 4. Kode lebih rapi
#
# 5. Mudah maintenance
#
# Prinsip OCP:
#
# Open for Extension
# = mudah tambah fitur baru
#
# Closed for Modification
# = kode lama tidak perlu diubah
#
# =========================================================
```

```python
#BAGIAN L
from abc import ABC, abstractmethod


# BASE CLASS UMUM
class Dokter(ABC):
    def _init_(self, nama):
        self.nama = nama

    @abstractmethod
    def konsultasi(self):
        pass


# KHUSUS DOKTER YANG BISA OPERASI
class DokterBisaOperasi(ABC):
    @abstractmethod
    def operasi(self):
        pass


class DokterAnak(Dokter, DokterBisaOperasi):
    def konsultasi(self):
        print(f"Dokter {self.nama} (Anak) melakukan konsultasi.")

    def operasi(self):
        print(f"Dokter {self.nama} melakukan operasi anak.")


class DokterGigi(Dokter, DokterBisaOperasi):
    def konsultasi(self):
        print(f"Dokter {self.nama} (Gigi) melakukan konsultasi.")

    def operasi(self):
        print(f"Dokter {self.nama} melakukan operasi gigi.")


class DokterKulit(Dokter, DokterBisaOperasi):
    def konsultasi(self):
        print(f"Dokter {self.nama} (Kulit) melakukan konsultasi.")

    def operasi(self):
        print(f"Dokter {self.nama} melakukan tindakan bedah kulit.")


class Psikiater(Dokter):
    # ✔️ Tidak dipaksa punya operasi
    def konsultasi(self):
        print(f"Dokter {self.nama} (Psikiater) melakukan terapi kejiwaan.")


def jadwal_konsultasi(dokter: Dokter):
    dokter.konsultasi()


def jadwal_operasi(dokter: DokterBisaOperasi):
    dokter.operasi()



# DEMO
dr_anak = DokterAnak("Andi")
dr_gigi = DokterGigi("Budi")
dr_kulit = DokterKulit("Citra")
dr_psikiater = Psikiater("Dewi")

#  Semua bisa konsultasi
jadwal_konsultasi(dr_anak)
jadwal_konsultasi(dr_psikiater)

print("-" * 30)

#  Hanya dokter yang memang bisa operasi
jadwal_operasi(dr_anak)
jadwal_operasi(dr_gigi)
jadwal_operasi(dr_kulit)

#  Tidak bisa (dan ini benar!)
# jadwal_operasi(dr_psikiater)
```

```python
#BAGIAN D
from abc import ABC, abstractmethod

# --- CONTOH BAIK (MENERAPKAN DIP) ---

# 1. ABSTRAKSI (Interface)
class MetodePengobatan(ABC):
    """
    Interface/Abstraksi.
    Semua jenis pengobatan harus memiliki fungsi 'eksekusi'.
    """
    @abstractmethod
    def eksekusi(self):
        pass

# 2. IMPLEMENTASI KONKRET (Low-level Modules)
class PemberianObat(MetodePengobatan):
    def eksekusi(self):
        return "Memberikan resep obat sesuai dosis."

class TerapiFisik(MetodePengobatan):
    def eksekusi(self):
        return "Melakukan sesi latihan fisik rehabilitasi."

class OperasiBedah(MetodePengobatan):
    def eksekusi(self):
        return "Melakukan tindakan pembedahan di ruang operasi."

# 3. KELAS TINGKAT TINGGI (High-level Module)
class Dokter:
    """
    Kelas Dokter sekarang bersifat umum.
    Ia hanya tahu bahwa ia harus menjalankan 'MetodePengobatan'.
    """
    def __init__(self, metode: MetodePengobatan):
        # Dependency Injection: Metode dipilih saat objek Dokter dibuat
        self.metode = metode

    def rawat_pasien(self):
        # Dokter hanya memanggil fungsi dari abstraksi
        tindakan = self.metode.eksekusi()
        print(f"Laporan Dokter: {tindakan}")

# --- PENGGUNAAN DI GOOGLE COLAB ---

# Dokter spesialis penyakit dalam menggunakan obat
dokter_internis = Dokter(PemberianObat())
dokter_internis.rawat_pasien()

# Dokter spesialis rehabilitasi menggunakan terapi fisik
dokter_fisio = Dokter(TerapiFisik())
dokter_fisio.rawat_pasien()

# Dokter bedah melakukan operasi
dokter_bedah = Dokter(OperasiBedah())
dokter_bedah.rawat_pasien()
```
